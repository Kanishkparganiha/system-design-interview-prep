# Exercise 02 — Inheritance & Polymorphism

## Concepts

| Term | What it means |
|------|---------------|
| **Inheritance** | A child class gets all attributes & methods of its parent |
| **Method Overriding** | The child replaces a parent’s method with its own |
| **Polymorphism** | Same method name, different behavior depending on the class |
| **`super()`** | Calls the parent class’s method from inside the child |
| **`isinstance()`** | Checks if an object is an instance of a class |

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement speak()")

class Cat(Animal):           # Cat inherits from Animal
    def speak(self):          # overrides parent’s speak()
        return f"{self.name} says Meow!"

class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

# Polymorphism in action:
animals = [Cat("Luna"), Dog("Buddy")]
for a in animals:
    print(a.speak())  # same method call, different output
```

## Bonus: Method Resolution Order (MRO)

Python supports **multiple inheritance**. When a class inherits from two parents, Python uses the **C3 linearization** algorithm to decide which method to call.

```python
class A:
    def greet(self): return "A"

class B(A):
    def greet(self): return "B"

class C(A):
    def greet(self): return "C"

class D(B, C):  # inherits from both B and C
    pass

print(D().greet())  # What does this print?
print(D.__mro__)    # Shows the lookup order
```

Try it below:

````python
# Try the MRO example here

class A:
    def greet(self): return "A"

class B(A):
    def greet(self): return "B"

class C(A):
    def greet(self): return "C"

class D(B, C):
    pass

print("D().greet() =", D().greet())
print("MRO:", [cls.__name__ for cls in D.__mro__])
````
