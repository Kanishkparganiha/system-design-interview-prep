# Exercise 03 — Encapsulation & Properties

## Concepts

| Term | What it means |
|------|---------------|
| **Encapsulation** | Hiding internal state, exposing only a controlled interface |
| **`_var`** (single underscore) | Convention: "private by agreement" — don't touch from outside |
| **`__var`** (double underscore) | Name-mangled to `_ClassName__var` — harder to access |
| **`@property`** | Turns a method into a read-only attribute |
| **`@attr.setter`** | Adds validation logic when setting an attribute |

### Why use `@property`?

Instead of writing Java-style getters/setters:
```python
# BAD — un-Pythonic
class User:
    def get_email(self): return self._email
    def set_email(self, val): self._email = val
```

Python uses `@property` for clean syntax:
```python
# GOOD — Pythonic
class User:
    def __init__(self, email):
        self._email = email   # store privately

    @property
    def email(self):          # getter
        return self._email

    @email.setter
    def email(self, value):   # setter with validation
        if "@" not in value:
            raise ValueError("Invalid email")
        self._email = value

u = User("a@b.com")
print(u.email)      # uses getter  → "a@b.com"
u.email = "x@y.com" # uses setter  → validates then sets
```
