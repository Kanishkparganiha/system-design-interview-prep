# Exercise 01 — Classes & Objects

## Concepts

| Term | What it means |
|------|---------------|
| **Class** | A blueprint/template for creating objects |
| **Object** | A concrete instance built from a class |
| **`__init__`** | The constructor — runs when you create an object |
| **`self`** | Refers to the current instance (like `this` in JS/TS) |
| **Instance attribute** | Data unique to each object (`self.name`) |
| **Class attribute** | Data shared by ALL instances (defined outside `__init__`) |

```python
class Dog:
    species = "Canis familiaris"  # class attribute (shared)

    def __init__(self, name, age):
        self.name = name  # instance attribute (unique)
        self.age = age

    def speak(self):
        return f"{self.name} says Woof!"

buddy = Dog("Buddy", 3)   # __init__ is called here
print(buddy.speak())       # "Buddy says Woof!"
print(buddy.species)       # "Canis familiaris"
```

## Task: Implement `BankAccount`

Create a class with this behavior:

```python
acc = BankAccount("Alice", 100.0)
acc.owner        # "Alice"
acc.balance      # 100.0
acc.deposit(50)  # balance becomes 150.0
acc.withdraw(30) # balance becomes 120.0, returns True
acc.withdraw(200)# balance stays 120.0, returns False (insufficient funds)
str(acc)         # "BankAccount(owner='Alice', balance=120.0)"
```

**Rules:**
1. `deposit(amount)` should raise `ValueError` if amount <= 0
2. `withdraw(amount)` should return `False` (not raise) if insufficient funds
3. Implement `__str__` for readable representation
4. Default balance should be `0.0` if not provided

````python
# YOUR CODE HERE

class BankAccount:
    #pass  # Replace with your implementation
    def __init__(self, owner:str, balance: float = 0.0):
        self.owner = owner
        self.balance = balance


    def withdraw(self,amount):
        if self.balance - amount <= 0:
            return False
        else:
            self.balance = self.balance - amount
            return True

    def deposit(self,amount):
        if amount <= 0:
            raise ValueError
        self.balance = self.balance + amount
        return self.balance
    
    def __str__(self):
        return f"BankAccount(owner='{self.owner}', balance={self.balance})"

````
