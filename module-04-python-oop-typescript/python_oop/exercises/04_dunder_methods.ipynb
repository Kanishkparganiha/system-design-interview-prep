# Exercise 04 — Dunder Methods & Operator Overloading

## Concepts

"Dunder" = **d**ouble **under**score. These special methods let your objects behave like built-in types.

| Dunder | Triggered by | Example |
|--------|--------------|--------|
| `__init__` | `MyClass()` | Constructor |
| `__str__` | `str(obj)`, `print(obj)` | Human-readable string |
| `__repr__` | `repr(obj)`, REPL display | Unambiguous string |
| `__len__` | `len(obj)` | Length |
| `__getitem__` | `obj[key]` | Indexing |
| `__eq__` | `obj1 == obj2` | Equality |
| `__lt__` | `obj1 < obj2` | Less than (enables `sorted()`) |
| `__add__` | `obj1 + obj2` | Addition |
| `__contains__` | `item in obj` | Membership |
| `__iter__` | `for x in obj` | Iteration |

```python
class Vector:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

v = Vector(1, 2) + Vector(3, 4)  # __add__ is called
print(v)  # Vector(4, 6)         # __repr__ is called
```
