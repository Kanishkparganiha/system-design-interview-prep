# 03 - Sliding Window

A sliding window maintains a contiguous subarray (or substring) and expands/shrinks it based on a condition.
It avoids recomputing from scratch each step — turning O(n²) into O(n).

**Pattern:** fixed-size window vs. dynamic window (shrink when condition violated)

**Problems in this notebook:**
1. Best Time to Buy and Sell Stock *(Easy)*
2. Longest Substring Without Repeating Characters *(Medium)*
3. Longest Repeating Character Replacement *(Medium)*
4. Minimum Window Substring *(Hard)*
5. Sliding Window Maximum *(Hard)*
