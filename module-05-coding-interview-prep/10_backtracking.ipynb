# 10 - Backtracking

Backtracking = DFS + pruning. Build a solution incrementally, abandon it (backtrack) as soon as you determine it can't lead to a valid answer.

**Template:**
```python
def backtrack(state):
    if is_solution(state):
        results.append(state[:])
        return
    for choice in choices(state):
        make_choice(choice)
        backtrack(state)
        undo_choice(choice)  # backtrack
```
