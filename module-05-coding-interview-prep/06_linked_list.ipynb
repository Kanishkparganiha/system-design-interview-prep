````python
# Shared ListNode definition — run this cell first
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def make_list(values):
    """Helper: create linked list from a Python list."""
    dummy = ListNode(0)
    cur = dummy
    for v in values:
        cur.next = ListNode(v)
        cur = cur.next
    return dummy.next

def to_list(node):
    """Helper: convert linked list to Python list."""
    result = []
    while node:
        result.append(node.val)
        node = node.next
    return result

print("ListNode defined. Run your solution cells below.")
````
