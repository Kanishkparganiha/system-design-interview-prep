# 09 - Dynamic Programming

DP = recursion + memoization (top-down) OR building a table of subproblems (bottom-up).

**When to use DP:** overlapping subproblems + optimal substructure (optimal solution built from optimal sub-solutions).

**Pattern:** define the state, write the recurrence, handle base cases.
