# 13 - Bit Manipulation

Bit tricks let you solve certain problems in O(1) space with no arithmetic loops.

**Key operators in Python:**
```python
n & 1       # check if odd (last bit is 1)
n >> 1      # right shift (divide by 2)
n << 1      # left shift (multiply by 2)
a ^ b       # XOR: 1 if bits differ, 0 if same
a & b       # AND
a | b       # OR
~a          # NOT (flips all bits)
n & (n-1)   # clear the lowest set bit
```
