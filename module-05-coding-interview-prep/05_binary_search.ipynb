# 05 - Binary Search

Binary search works on **sorted** data (or any data with a monotonic property). It eliminates half the search space each iteration — O(log n) instead of O(n).

The key insight: binary search isn't just for finding a value — it can search for the **smallest/largest value satisfying a condition**.
