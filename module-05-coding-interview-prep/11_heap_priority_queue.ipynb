# 11 - Heap / Priority Queue

A heap lets you efficiently get the min or max element in O(1) and insert/remove in O(log n).

Python's `heapq` is a **min-heap**. For a max-heap, negate values.

```python
import heapq
h = []
heapq.heappush(h, val)     # push
heapq.heappop(h)           # pop minimum
heapq.heapify(lst)         # convert list to heap in-place, O(n)
```
